```python
import os\nimport io\nimport base64\nimport threading\nfrom typing import Optional\n\nimport torch\nfrom fastapi import FastAPI, Header, HTTPException\nfrom pydantic import BaseModel, Field\nfrom PIL import Image\nfrom pyngrok import ngrok\nimport nest_asyncio\nimport uvicorn\nfrom diffusers import DiffusionPipeline
```

```python
HF_TOKEN = os.getenv('HF_TOKEN', '')\nAPI_TOKEN = os.getenv('COLAB_API_TOKEN', '')\n\npipe = DiffusionPipeline.from_pretrained(\n    'stabilityai/stable-diffusion-xl-base-1.0',\n    torch_dtype=torch.float16,\n    use_safetensors=True,\n    token=HF_TOKEN if HF_TOKEN else None,\n)\npipe = pipe.to('cuda')\npipe.enable_attention_slicing()
```

```python
class GenerateRequest(BaseModel):\n    prompt: str = Field(..., min_length=1, max_length=2000)\n    negative_prompt: str = ''\n    image_size: int = 1024\n    num_images: int = 1\n\napp = FastAPI(title='SDXL Colab API')\n\ndef pil_to_base64(img: Image.Image) -> str:\n    buf = io.BytesIO()\n    img.save(buf, format='PNG')\n    return base64.b64encode(buf.getvalue()).decode('utf-8')\n\n@app.post('/generate')\ndef generate(payload: GenerateRequest, authorization: Optional[str] = Header(default=None)):\n    if API_TOKEN:\n        expected = f'Bearer {API_TOKEN}'\n        if authorization != expected:\n            raise HTTPException(status_code=401, detail='Unauthorized')\n\n    if payload.image_size not in [512, 768, 1024]:\n        raise HTTPException(status_code=400, detail='image_size must be 512, 768, or 1024')\n\n    if payload.num_images < 1 or payload.num_images > 4:\n        raise HTTPException(status_code=400, detail='num_images must be between 1 and 4')\n\n    result = pipe(\n        prompt=payload.prompt,\n        negative_prompt=payload.negative_prompt,\n        height=payload.image_size,\n        width=payload.image_size,\n        num_images_per_prompt=payload.num_images,\n        guidance_scale=7.5,\n        num_inference_steps=35,\n    )\n\n    images_b64 = [pil_to_base64(img) for img in result.images]\n    return {'images': images_b64}
```

```python
nest_asyncio.apply()\nNGROK_AUTHTOKEN = os.getenv('NGROK_AUTHTOKEN', '')\nif NGROK_AUTHTOKEN:\n    ngrok.set_auth_token(NGROK_AUTHTOKEN)\n\npublic_url = ngrok.connect(8000).public_url\nprint('Public API URL:', public_url + '/generate')\n\nconfig = uvicorn.Config(app, host='0.0.0.0', port=8000, log_level='info')\nserver = uvicorn.Server(config)\nthread = threading.Thread(target=server.run)\nthread.start()
```
